# Imports and Reading Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/cost-of-living-index-by-country-by-number-2024/Cost_of_Living_Index_by_Country_2024.csv')
df.head()
```

# Data Understanding

```python
df.info()
```

```python
df.describe()
```

## Data Preprocessing

```python
top10_D = df.sort_values(by='Cost of Living Index', ascending= False).head(10)
top10_A = df.sort_values(by='Cost of Living Index', ascending= True).head(10)
```

```python
RentA = df.sort_values(by='Rent Index', ascending= True).head(10)
RentD = df.sort_values(by='Rent Index', ascending= False).head(10)
```

```python
GroceriesA = df.sort_values(by='Groceries Index', ascending= True).head(10)
GroceriesD = df.sort_values(by='Groceries Index', ascending= False).head(10)
```

```python
RestaurantPriceA = df.sort_values(by='Restaurant Price Index', ascending= True).head(10)
RestaurantPriceD = df.sort_values(by='Restaurant Price Index', ascending= False).head(10)
```

```python
LPP_A = df.sort_values(by='Local Purchasing Power Index', ascending= True).head(10)
LPP_D = df.sort_values(by='Local Purchasing Power Index', ascending= False).head(10)
```

## Visualization the Data

```python
sns.set_theme(style="ticks", rc={'figure.figsize': (15, 7)})
plt.subplot(1, 2, 1)
sns.barplot(x='Cost of Living Index', y='Country',data= top10_D, palette='icefire')
plt.title('Top 10 Most Expensive Countries')
plt.subplot(1, 2, 2)
sns.barplot(x='Cost of Living Index', y='Country',data= top10_A, palette='icefire')
plt.title('Top 10 Most Inexpensive Countries')
plt.show()
```

```python
plt.subplot(1, 2, 1)
sns.barplot(x='Rent Index', y='Country',data= RentD, palette='icefire')
plt.title('Top 10 Most Expensive Countries by Rent')
plt.subplot(1, 2, 2)
sns.barplot(x='Rent Index', y='Country',data= RentA, palette='icefire')
plt.title('Top 10 Most Inexpensive Countries by Rent')
plt.show()
```

```python
plt.subplot(1, 2, 1)
sns.barplot(x='Groceries Index', y='Country',data= GroceriesD, palette='icefire')
plt.title('Top 10 Most Expensive Countries by Groceries')
plt.subplot(1, 2, 2)
sns.barplot(x='Groceries Index', y='Country',data= GroceriesA, palette='icefire')
plt.title('Top 10 Most Inexpensive Countries by Groceries')
plt.show()
```

```python
sns.set_theme(style="ticks", rc={'figure.figsize': (15, 7)})
plt.subplot(1, 2, 1)
sns.barplot(x='Local Purchasing Power Index', y='Country',data= LPP_D, palette='icefire')
plt.title('Top 10 Highest Local Purchasing Power Index Countries')
plt.subplot(1, 2, 2)
sns.barplot(x='Local Purchasing Power Index', y='Country',data= LPP_A, palette='icefire')
plt.title('Top 10 Lowest Local Purchasing Power Index Countries')
plt.show()
```

```python
sns.set_theme(style="ticks", rc={'figure.figsize': (15, 7)})
plt.subplot(1, 2, 1)
sns.barplot(x='Restaurant Price Index', y='Country',data= RestaurantPriceD, palette='icefire')
plt.title('Top 10 Highest Restaurant Price Index Countries')
plt.subplot(1, 2, 2)
sns.barplot(x='Restaurant Price Index', y='Country',data= RestaurantPriceA, palette='icefire')
plt.title('Top 10 Lowest Restaurant Price Index Countries')
plt.show()
```

```python
sns.set_theme(style="ticks", rc={'figure.figsize': (15, 7)})
df_corr = df[['Cost of Living Index','Rent Index','Groceries Index','Restaurant Price Index']].corr()
sns.heatmap(df_corr,annot= True, cmap="Greens",)
plt.show()
```

```python
sns.set_theme(style="ticks", rc={'figure.figsize': (15, 7)})
sns.pairplot(
    df, vars=["Cost of Living Index", "Rent Index", "Groceries Index", "Restaurant Price Index"]
)
plt.show()
```
